```python
new_df = df.replace([-99999, -88888, 0], np.nan)
new_df
```

```python
new_df = df.replace({'temperature': -99999,
                     'windspeed': [-99999, -88888],
                     'event': '0'}, np.nan)
new_df
```

```python
df = pd.read_csv('Weather_Data_cb_str.csv')
df
```

```python
new_df = df.replace('[a-zA-Z]','', regex= True)
new_df
```

```python
new_df = df.replace({'temperature': '[a-zA-Z]',
                     'windspeed': '[a-zA-Z]'}, '', regex= True)
new_df
```

```python
new_df = df.replace({'temperature': {'[a-zA-Z]':''},
                     'windspeed': {'[a-zA-Z]':''}, 
                     'event': {'0': 'No Event'}}, regex= True)
new_df
```

```python
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
df['score'] = df['score'].replace(['poor', 'average', 'good', 'exceptional'], [1, 2, 3, 4])
df
```

```python
mapping = {'poor': 1, 'average': 2, 'good': 3, 'exceptional': 4}
df['score'] = df['score'].replace(mapping)
df
```
