```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
df = pd.concat([india_weather, us_weather], ignore_index= True)
df
```

```python
df = pd.concat([india_weather, us_weather], keys = ["india", "us"])
df
```

```python
df.loc["india"]
```

```python
temp_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30]
})
temp_df
```

```python
humidity_df = india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "humidity": [80, 60, 78]
})
humidity_df
```

```python
india_weather = pd.concat([temp_df, humidity_df], axis = 1)
india_weather
```

```python
india_weather = pd.concat([temp_df, humidity_df['humidity']], axis = 1)
india_weather
```
