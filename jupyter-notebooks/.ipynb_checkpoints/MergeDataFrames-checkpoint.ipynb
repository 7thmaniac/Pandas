```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2, on='city')
df3
```

```python
df4 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,75],
})
df5
```

```python
df6 = pd.merge(df4, df5, on='city')
df6
```

```python
df7 = pd.merge(df4, df5, on='city', how='left')
df7
```

```python
df8 = pd.merge(df4, df5, on='city', how='right')
df8
```

```python
df9 = pd.merge(df4, df5, on='city', how='inner')
df9
```

```python
df10 = pd.merge(df4, df5, on='city', how='outer')
df10
```

## Suffixes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city', suffixes=('_df1', '_df2'))
df3
```

```python
df4 = pd.merge(df1, df2, on='city', suffixes=('_df1', '_df2'), how='outer')
df4
```
