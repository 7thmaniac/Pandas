```python
python_dictionary = Weather_Data
dict_df = pd.DataFrame(python_dictionary)
dict_df
```

## Using Tuples

```python
Weather_Data = [
    ('1/1/2017', 32, 6, 'Rain'),
    ('1/2/2017', 35, 7, 'Sunny'),
    ('1/3/2017', 28, 2, 'Snow')
]
tuple_df = pd.DataFrame(Weather_Data, columns=['Date', 'Temperature', 'Wind_Speed', 'Event'])
tuple_df
```

## Using list of dictionaries

```python
Weather_Data = [
{ 'Date': '1/1/2017', 'Temperature': 32, 'Wind_Speed': 6, 'Event': 'Rain'},
{ 'Date': '1/2/2017', 'Temperature': 35, 'Wind_Speed': 7, 'Event': 'Sunny'},
{ 'Date': '1/3/2017', 'Temperature': 28, 'Wind_Speed': 2, 'Event': 'Snow'}
]
lst_of_dict_df = pd.DataFrame(Weather_Data)
lst_of_dict_df
```

# Reading And Writing Different Types Of Files Using Pandas

## Reading Using CSV

```python
csv_df = pd.read_csv("Stock_Data.csv")
csv_df
```

#### Here, there is a cell containing data in the first row. While reading pandas usually takes first row as headers to specify from where our data need to be read or omit unnecessary cells we have to specify head

```python
csv_df = pd.read_csv("Stock_Data.csv", skiprows = 1)
csv_df
```

#### The above skiprows option skipped the first row, it generally skips how ever many rows specified

```python
csv_df_skp = pd.read_csv("Stock_Data.csv", skiprows = 3)
csv_df_skp
```

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1)
csv_df
```

#### header option specifies which row to take as header

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1)
csv_df
```

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 3)
csv_df
```

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1, nrows = 3)
csv_df
```

#### We can fill na values with NaN while reading file

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1, nrows = 3, na_values = ['not available', 'n.a.'])
csv_df
```

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1, na_values = ['not available', 'n.a.'])
csv_df
```

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1, na_values = ['not available', 'n.a.',-1])
csv_df
```

#### As you can see in revenue column there cannot be negative values so I wanted to replace -1 with NaN, but the code also changed the -1.00 in eps, where eps can be negative, so the alternatives is we can the column names as dictionaries as shown below selectively

```python
csv_df = pd.read_csv("Stock_Data.csv", header = 1, na_values = {'eps' : ['not available', 'n.a.'],
                                                                'revenue' : ['not available', 'n.a.',-1],
                                                                'price' : ['not available', 'n.a.'],
                                                                'people' : ['not available', 'n.a.']})
csv_df
```

## Writing to CSV

```python
csv_df.to_csv('New_Stock_Data.csv')
new_csv_df = pd.read_csv('New_Stock_Data.csv')
new_csv_df
```

#### Removing unnecessary index column

```python
csv_df.to_csv('New_Stock_Data.csv', index = False)
new_csv_df = pd.read_csv('New_Stock_Data.csv')
new_csv_df
```
